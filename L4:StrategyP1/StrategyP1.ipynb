```python
import backtrader as bt
import pandas as pd
import yfinance as yf
```

```python
tesla = yf.download('TSLA', start='2020-01-01')
tesla.info()
```

```python
## PARSE THE DATA
data = bt.feeds.PandasData(dataname=tesla)
```

```python
class TestSTG(bt.Strategy):
    def __init__(self):
        pass
    def next(self):
        pass

cerebro = bt.Cerebro()
cerebro.adddata(data)
cerebro.addstrategy(TestSTG)
cerebro.run()
```

```python
class TestClose(bt.Strategy):
    def __init__(self):
        pass
    def next(self):
        dataclose = self.datas[0].close[0]
        print(dataclose)


cerebro = bt.Cerebro()
cerebro.adddata(data)
cerebro.addstrategy(TestClose)
cerebro.run()
```

```python
class TestClosei(bt.Strategy):
    def __init__(self):
        self.dataclose = self.datas[0].close
    def next(self):
        print(self.dataclose[0])

cerebro = bt.Cerebro()
cerebro.adddata(data)
cerebro.addstrategy(TestClosei)
cerebro.run()
```

```python
class TestLog(bt.Strategy):
    def __init__(self):
        self.dataclose = self.datas[0].close

    def log(self, txt):
        dt = self.datas[0].datetime.datetime()
        print(f'{dt} | {txt}')

    def next(self):
        x = self.dataclose[0]
        self.log(txt=x)


cerebro = bt.Cerebro()
cerebro.adddata(data)
cerebro.addstrategy(TestLog)
cerebro.run()
```
